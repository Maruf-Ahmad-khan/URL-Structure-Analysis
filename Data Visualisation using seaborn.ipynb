# Find the count of the days

```python
def Count_Number_Days(tips):
     plt.figure(figsize=(6,3))
     sns.countplot(x = 'day', data = tips)
     plt.title("Count then number of Days")
     plt.show()
     
tips = sns.load_dataset("tips")
Count_Number_Days(tips)
```

# Find the count of time and smoker

```python
import matplotlib.pyplot as plt
import seaborn as sns

def Count_time(tips):
    plt.figure(figsize=(6,3))  # Correctly set the figure size
    sns.countplot(x='time', data=tips, hue='smoker')
    plt.title("Count of the Days")
    plt.show()

tips = sns.load_dataset("tips")
Count_time(tips)  # Correct way to call the function
```

```python
def Count_Smokers(tips):
     plt.figure(figsize=(6,3))
     sns.countplot(x = 'smoker', data = tips)
     plt.title("Count of Smokers")
     plt.show()
     
tips = sns.load_dataset("tips")
Count_Smokers(tips)
```

# Count of Male and Female

```python
def Gender_Count(tips):
     plt.figure(figsize=(6,3))
     sns.countplot(x = 'sex', data = tips)
     plt.title("Count of Gender")
     plt.show()
     
tips = sns.load_dataset("tips")
Gender_Count(tips)
```

# Establish the relationship using relplot between two numerical columns

```python
def Scatter_Plot(tips):
     plt.figure(figsize=(6,3))
     sns.relplot(data = tips, x = "total_bill", y = 'tip', hue = "time")
     plt.title("Relationship between tips and bils columns")
     plt.show()


tips = sns.load_dataset("tips")
Scatter_Plot(tips)     
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def Scatter_Plot(tips):
    plt.figure(figsize=(6, 3))
    sns.scatterplot(data=tips, x="total_bill", y='tip')
    plt.title("Relationship between tips and total bill columns")
    plt.show()

tips = sns.load_dataset("tips")
Scatter_Plot(tips)
```

```python
def Line_Plot(tips):
     plt.figure(figsize=(6,3))
     sns.relplot(data = tips, x = 'day', y = 'total_bill', kind='line', errorbar = None)
     plt.title("Line plot between day and total_bill")
     plt.show()
     
tips = sns.load_dataset('tips')
Line_Plot(tips)
```

```python
def Line_Plot(tips):
     plt.figure(figsize=(6,3))
     sns.relplot(data = tips, x = 'day', y = 'total_bill', kind='line', estimator = None)
     plt.title("Line plot between day and total_bill")
     plt.show()
     
tips = sns.load_dataset('tips')
Line_Plot(tips)
```
